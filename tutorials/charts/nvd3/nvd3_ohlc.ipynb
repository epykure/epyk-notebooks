# NVD3 


## OHLC Chart

More information about this chart [here](http://nvd3.org/)

```python
from epyk.core.Page import Report
from epyk.tests import data_urls

rptObj = Report()


data_rest = rptObj.py.requests.csv(data_urls.PLOTLY_APPLE_PRICES)

ohlc = rptObj.ui.charts.nvd3.ohlc(data_rest, closes=["AAPL.Close"], highs=["AAPL.High"], lows=["AAPL.Low"], opens=["AAPL.Open"], x_axis='Date')

rptObj.outs.jupyter()
```

For more details about this charts and the various options and properties available please refer to the [official website](http://nvd3.org/)

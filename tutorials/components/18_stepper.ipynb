# Stepper

Stepper [component](00_components.ipynb) creation in Epyk will be using the function **stepper** from the ui property.
<i>
the ui entry point will give you access to all the registered components.
Different flavour of buttons are available in the framework.</i>

For more custom examples please have a look at the [Custom section](../custom/stepper.ipynb)


## Basic stepper

```python
from epyk.core.Page import Report

rptObj = Report()

records = []
for i in range(6):
  records.append({'status': 'success', "tooltip": "test", 'value': 'test', 'title': 'Ok %s' % i},)
sp = rptObj.ui.stepper(records)


rptObj.outs.jupyter()
```

```python
from epyk.core.Page import Report

rptObj = Report()

sp_a = rptObj.ui.steppers.arrow(records)


rptObj.outs.jupyter()
```

```python
from epyk.core.Page import Report

rptObj = Report()

sp_a = rptObj.ui.steppers.vertical(records, shape='rectangle')

rptObj.outs.jupyter()
```

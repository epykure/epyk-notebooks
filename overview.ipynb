# Project

Implementation of Epyk started from a need for Python developers to be more involved in the entire application design. As data transformation is key in each product it was important for them to be involved in the final display to ensure a simple and efficient data processing.

The need was about using Python to do the data transformation but also the template for the display. A simple prototype using Flask and Jinja proved the feasability but we did not want to change languages... this was the beginning of Epyk.

Today Epyk is including multiple Javascript libraries in order to simplify as much as possible the Front End implementation. It will use its rich ecosystem of components plus its internal Javascript tranpiler to render a rich HTML 5 page.

Components and JavaScript features are added following request. Do not hesitate to share your needs and we will add this to our roadmap !


## Installation


Epyk can be installed easily to an Jupyter Notebook by running the below command 

> !pip install epyk

This will get the latest version of the package to the Pypi repository

Once installed it will be possible to prototype and create part of components which can then be reused in your final application. Jupyter is great platform to share code but also to advertise your work but this code should be then used to production with a minimum of changes. In the same way that Codepen is a very efficient plateform to share front end code.

This is what this framework should allow you to to both and according to your environment the output generated will be different.

Today Epyk ecosystem is using the best of the Javascript and Python ecosystem to allow you to be a full stack developer from Python.

So have a look at the tutorials and test it online directly !

## General rule

Any changes or fixes in any Javascript library should be temporary and the issue should be raised to the Javascript framework in order to have a centralised and permanent fix. In the same way it helps Python developer, Epyk framework should benefit to the Javascript community to improve its underlying modules !

Same thing, any improvment on the underlying JavaScript modules should be raised directly on the module repositories. An example in Codepen could help explaining the issue.
